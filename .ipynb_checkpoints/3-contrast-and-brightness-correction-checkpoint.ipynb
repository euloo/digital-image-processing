```python
from skimage.io import imread, imsave, imshow
from skimage    import img_as_float
from skimage    import img_as_ubyte
from numpy      import array_equal
from numpy      import clip, sort
from numpy      import dstack

img = imread("tiger-color.png")
img2 = imread("tiger-stable-contrast.png")
#img = imread('img.png')

img_f = img_as_float(img)

R = img_f[:,:,0]
G = img_f[:,:,1]
B = img_f[:,:,2]

Y =  0.2126* R + 0.7152* G + 0.0722* B
U = -0.0999* R - 0.3360* G + 0.4360* B
V =  0.6150* R - 0.5586* G - 0.0563* B


img_sort_list = sort(Y, axis=None)
k = round(img_sort_list.size*0.05)
x_min, x_max = img_sort_list[k], img_sort_list [-k - 1]


Y = clip((Y - x_min)/(x_max - x_min), 0 , 1)


R = clip(Y + 1.2803*V, 0 , 1)
G = clip(Y - 0.2148*U - 0.3805*V, 0 , 1)
B = clip(Y + 2.1279*U, 0 , 1)


img_combined = img_as_ubyte(dstack((R , G, B)))


#imsave ('out_img.png', img_combined)
imshow(img_combined)
array_equal(img2, img_combined)
```

```python
from skimage.io import imread, imsave, imshow
from skimage    import img_as_float
from skimage    import img_as_ubyte
from numpy      import array_equal
from numpy      import clip, sort
from numpy      import dstack, mean

img = imread("railroad.png")
img2 = imread("railroad-gray-world.png")
img = img_as_float(img)

R = img[:,:, 0]#    
G = img[:,:, 1]
B = img[:,:, 2]

avg = (R.mean() + G.mean() + B.mean())/3
rw = R.mean()/avg
gw = G.mean()/avg
bw = B.mean()/avg

R = clip(R/rw, 0, 1)
G = clip(G/gw, 0, 1)
B = clip(B/bw, 0, 1)

img_combined = img_as_ubyte(dstack((R , G, B)))


#imsave ('out_img.png', img_combined)
imshow(img_combined)
array_equal(img2, img_combined)

```
