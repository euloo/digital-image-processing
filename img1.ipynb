```python
x1, x2, x3, x4 =  0, 0, 0, 0


for i in range(col):
    if all(c!= img2[xs, i]):
        break
    x1 += 1
    
for i in range(row):
    if all(c!= img2[i, ys]):
        break
    x2 += 1
    
for i in range(row):
    if all(c!= img2[xs, col - i]):
        break
    x3 += 1
    
for i in range(col):
    if all(c!= img2[row - i, ys]):
        break
    x4 += 1
    
print(x1, x2, x3, x4)
```

```python
from numpy import roll
(r* g).sum()
```

```python
cor = 0
for i in range(-20, 20, 1):
    bi = roll(b, i, 0)
    for j in range(-20, 20, 1):
        bj = roll(bi, j, 1)
        cb = (bj * g).sum()
        if cb>cor:
            cor = cb
            print(cor)
            s0 = i
            s1 = j
rolled1 = roll(b, s0, 0)
rolled2 = roll(rolled1, s1, 1)
cor
```

```python
cor = 0
for i in range(-20, 20, 1):
    ri = roll(r, i, 0)
    for j in range(-20, 20, 1):
        rj = roll(ri, j, 1)
        cb = (rj * g).sum()
        if cb>cor:
            cor = cb
            s0 = i
            s1 = j
rolled3 = roll(r, s0, 0)
rolled4 = roll(rolled3, s1, 1)
cor
```
